```python
#read price data from csv
url = "https://raw.githubusercontent.com/PhaNtazM1337/CTC22/main/Price%20Data.csv"
raw_data = pd.read_csv(url, header=0, index_col=0)

#build stock tickers/symbols list
assets = list(raw_data)
#Initialize randomized weights/fractional holdings
weights = np.random.rand(40)
weights = weights/np.sum(weights)
raw_data.head()
```

```python
#create dataframe to store close price of stocks
df = raw_data.drop(labels=["Basic shares outstanding"])
df
```

```python
#create dataframe to store close price of stocks
returns = df.pct_change()
returns
```

```python
#Create and show the covariance matrix
cov_matrix_annual = df.pct_change().apply(lambda x: np.log(1+x)).cov() * 252
cov_matrix_annual
```

```python
port_variance = np.dot(weights.T, np.dot(cov_matrix_annual, weights))
port_variance
```

```python
port_volatility = np.sqrt(port_variance)
port_volatility
```

```python
port_annual_return = np.sum(returns.mean() * weights) * 252
port_annual_return
```

```python
port_sharpe = port_annual_return / port_volatility
port_sharpe
```

```python
num_ports = 5000
all_weights = np.zeros((num_ports, 40))
ret_arr = np.zeros(num_ports)
vol_arr = np.zeros(num_ports)
sharpe_arr = np.zeros(num_ports)
log_return = np.log(df/df.shift(1))
for ind in range(num_ports): 
    # weights 
    weights = np.array(np.random.random(40)) 
    weights = weights/np.sum(weights)  
	
    # save the weights
    all_weights[ind,:] = weights
	
    # expected return 
    ret_arr[ind] = np.sum((log_return.mean()*weights)*252)

    # expected volatility 
    vol_arr[ind] = np.sqrt(np.dot(weights.T,np.dot(log_return.cov()*252, weights)))

    # Sharpe Ratio 
    sharpe_arr[ind] = ret_arr[ind]/vol_arr[ind]
```

```python
# plot the data
plt.figure(figsize=(12,8))
plt.scatter(vol_arr,ret_arr,c=sharpe_arr,cmap='plasma')
plt.colorbar(label='Sharpe Ratio')
plt.xlabel('Volatility')
plt.ylabel('Return')
```

```python
#Calculate CAPM: Expected Return = beta * (expected market return)


# Normalize stock data based on initial price
def normalize(df):
  x = df.copy()
  for i in x.columns[1:]:
    x[i] = x[i]/x[i][0]
  return x

normalize(df)

# Function to plot interactive plot
def interactive_plot(df, title):
  fig = px.line(title = title)
  for i in df.columns[1:]:
    fig.add_scatter(x = df['Ticker'], y = df[i], name = i)
  fig.show()

interactive_plot(normalize(df), 'Normalized Prices')
```

```python
#Daily Return function:
def daily_return(df):
  df_daily_return = df.copy()
  for i in df.columns[1:]:
      for j in range(1, len(df)):
          df_daily_return[i][j] = ((df[i][j]- df[i][j-1])/df[i][j-1]) * 100
      df_daily_return[i][0] = 0
  return df_daily_return

stocks_daily_return = daily_return(df)
stocks_daily_return

# Histogram of daily returns (in %)
stocks_daily_return.boxplot(figsize=(12, 10), grid=False)
plt.title("Daily returns of the stocks")
```

```python
# Initialize empty dictionaries to save results
beta,alpha = dict(), dict()
# Make a nice subplot
fig, axes = plt.subplots(1,3, dpi=150, figsize=(15,8))
axes = axes.flatten()
# Loop on every stock daily return
for idx, stock in enumerate(stocks_daily_return.columns):
# Ignoring the date and Market Columns 
    if stock != "Date" and stock != "Market":
        # scatter plot between stock and the Market
        stocks_daily_return.plot(kind = "scatter", x = "Market", y = stock, ax=axes[idx-1])
 
        # Fit a line (regression using polyfit of degree 1)
        b_, a_ = np.polyfit(stocks_daily_return[stock], stocks_daily_return["Market"], 1)
 
        regression_line = b_ * stocks_daily_return["Market"] + a_
        axes[idx-1].plot(stocks_daily_return["Market"], regression_line, "-", color = "r")
        # save the regression coeeficient for the current stock
        beta[stock] = b_
        alpha[stock] = a_
plt.suptitle("Beta estimation: regression between Market and individual stock daily performance", size=20)
```
